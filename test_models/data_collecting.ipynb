```python
def show_classes_examples(df : pd.core.frame.DataFrame, label_col_name : str, text_column : str, dict_label_and_meaning = None):
  all_labels = df[label_col_name].unique()
  all_labels.sort()
  for label in all_labels:
    if dict_label_and_meaning is None:
      name = label
    else:
      name = dict_label_and_meaning[label]
    print(f"----------\nLabel \"{name}\"\n")
    texts = df[df[label_col_name] == label][text_column].to_list()
    for i, text in enumerate(texts):
      print(f"{i+1}. {text}\n\n")
      if i == 4:
        break
  
```

## emotion train

```python
from datasets import load_dataset
import os.path

def get_emotion_train_as_csv() -> pd.core.frame.DataFrame:

    if  not os.path.exists(PATH_TO_TRAIN_EMOTION):
      emotion_train = load_dataset("emotion")
      emotion_train['train'].to_csv(PATH_TO_TRAIN_EMOTION, index = False)
    ds = pd.read_csv(PATH_TO_TRAIN_EMOTION)
    return ds
```

## emotion

```python
test_emotion = pd.read_csv(PATH_TO_TEST_EMOTION).dropna()
test_emotion
```

```python
show_classes_examples(test_emotion, 'category', 'clean_text', {-1.0 : "negative", 0.0 : "neutral", 1.0 : "positive"})
```

check that this data is different from the training

```python
emotion_test_texts = test_emotion.clean_text.to_list()
```

```python
list(set(emotion_train_texts) & set(emotion_test_texts))
```

Get smaller dataset to easier analysis

```python
test_emotion_neg = test_emotion[test_emotion['category'] == -1.0].head(500)
test_emotion_neut = test_emotion[test_emotion['category'] == 0.0].head(500)
test_emotion_pos = test_emotion[test_emotion['category'] == 1.0].head(500)
```

```python
test_emotion = test_emotion_neg.append(test_emotion_neut, ignore_index=True).append(test_emotion_pos, ignore_index=True).sample(frac=1)
test_emotion
```

```python
test_emotion = test_emotion.rename(columns = {"clean_text":"text"})
test_emotion
```

```python
pd.Series([-1.0, 0.0, 1.0, 0.0])
one_hot_cats = pd.get_dummies(test_emotion.category)
one_hot_cats
```

```python
test_emotion = test_emotion.join(one_hot_cats).drop(columns=['category']).rename(columns={-1.0:"is_negative", 0.0:"is_neutral", 1.0:"is_positive"})
test_emotion
```

```python
test_emotion.to_csv(PATH_TO_TEST_EMOTION, index = False)
```

## combine this ds

```python
test_spam = pd.read_csv(PATH_TO_TEST_SPAM)
test_spam
```

```python
test_spam['is_toxic'] = 0
test_spam['is_negative'] = 0
test_spam['is_neutral'] = 0
test_spam['is_positive'] = 0
test_spam
```

```python
test_toxic = pd.read_csv(PATH_TO_TEST_TOXIC)
test_toxic
```

```python
test_toxic['is_spam'] = 0
test_toxic['is_negative'] = 1
test_toxic['is_neutral'] = 0
test_toxic['is_positive'] = 0
test_toxic
```

```python
test_emotion = pd.read_csv(PATH_TO_TEST_EMOTION)
test_emotion
```

```python
test_emotion['is_spam'] = 0
test_emotion['is_toxic'] = 0
test_emotion
```

```python
full_ds = test_spam.append(test_toxic, ignore_index=True).append(test_emotion, ignore_index=True).sample(frac=1)
```

```python
full_ds
```
